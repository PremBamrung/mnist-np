## Param√®tres

```python
X_train = np.load("../data/number/numpy/train-images.npy").reshape(-1, 28*28)
Y_train = np.load("../data/number/numpy/train-labels.npy")

X_test = np.load("../data/number/numpy/t10k-images.npy").reshape(-1, 28*28)
Y_test = np.load("../data/number/numpy/t10k-labels.npy")
neurons = 256
input_dim = 28*28
ouput_dim = 10
learning_rate = 1e-3
batch_size = 2048
eps = 1e-8
```

## Initialisation des poids

```python
# Initialisation des poids
# W matrice
# b vecteur
# normalis√©, moyenne 0 √©cart-type 1
# Initialisation Xavier et He n'influencent que sur des r√©seaux tr√®s profonds

W = 0.01 * np.random.randn(input_dim, neurons)
b = np.zeros((1, neurons))
W2 = 0.01 * np.random.randn(neurons, ouput_dim)
b2 = np.zeros((1, ouput_dim))
```

## Cr√©ation des batchs

```python
def generate_batch(X, Y, batch_size):
    """
    Generate a batch for X and Y of corresponding indexes
    """
    batch = np.random.choice(X.shape[0], batch_size)
    return X[batch], Y[batch]


X_batch, Y_batch = generate_batch(X_train,
                                  Y_train,
                                  batch_size)

X_test_batch, Y_test_batch = generate_batch(X_test,
                                            Y_test,
                                            batch_size)
```

## Premi√®re couche

```python
def ponderation(X, W, b):
    """
    1 ponderation
    return : output layer
    """
    return np.dot(X, W)+b


ponderation(X_batch, W, b)
```

```python
print(X_batch.shape)
print(W.shape)
print(b.shape)

ponderation(X_batch, W, b).shape
```

***ReLu (rectified linear units)***

* It was found to greatly accelerate (e.g. a factor of 6 in Krizhevsky et al.) the convergence of stochastic gradient descent compared to the sigmoid/tanh functions. It is argued that this is due to its linear, non-saturating form.pourquoi relu


* Compared to tanh/sigmoid neurons that involve expensive operations (exponentials, etc.), the ReLU can be implemented by simply thresholding a matrix of activations at zero.

```python
def relu(x):
    """
    Relu activation function
    return : relu output (max(0,x))
    """
    return np.maximum(0, x)


hidden_layer = relu(ponderation(X_batch, W, b))
hidden_layer.shape
```

## Deuxi√®me couche

```python
scores = ponderation(hidden_layer, W2, b2)
scores.shape
```

\begin{equation}
softmax = \frac{e^{f_{y_i}}}{ \sum_j e^{f_j} }
\end{equation}

```python
def softmax(scores, eps):
    """
    Compute class probabilities by softmax
    return : softmax output (probabilities)
    """
    exp = np.exp(scores)
    probs = exp/(np.sum(exp, axis=1, keepdims=True)+eps)
    return probs


probs = softmax(scores, eps)
probs[0]
```

## Loss, accuracy

\begin{equation}
L_i = -\log\left(\frac{e^{f_{y_i}}}{ \sum_j e^{f_j} }\right)
\end{equation}

\begin{equation}
L =  \frac{1}{N} \sum_i L_i 
\end{equation}

```python
def cross_entropy(probs, y, num_examples, eps):
    """
    Compute the loss: average cross-entropy loss and regularization
    return : cross_entropy loss
    """
    correct_logprobs = -np.log(probs[range(num_examples), y])+eps
    data_loss = np.sum(correct_logprobs)/num_examples

    return data_loss


loss = cross_entropy(probs, Y_batch, batch_size, eps)
loss
```

```python
def get_accuracy(probs, y):
    """
    Compute accuracy after ponderation
    Do a comparaison between the predicted value and the true value
    return : accuracy (from 0 to 1)
    """
    predict_class = np.argmax(probs, axis=1)
    return np.mean(predict_class == y)


get_accuracy(probs, Y_batch)
```

## Gradient 

\begin{equation}
p_k = \frac{e^{f_k}}{ \sum_j e^{f_j} } \hspace{1in} L_i =-\log\left(p_{y_i}\right)
\end{equation}


\begin{equation}
\frac{\partial L_i }{ \partial f_k } = p_k - \mathbb{1}(y_i = k)
\end{equation}

$p_k$ : softmax pour un cas

$L_i$ : Loss pour un cas

$f_i$ : la fonction qui repr√©sente le r√©seau de neuronne 


On veut savoir comment la sortie de $f$ doit changer pour diminuer $L_i$ dans cet exemple. On peut d√©rive le gradient $\partial L_i / \partial f_k$. La loss $L_i$ est d√©duite de $p$ qui est d√©duit de $f$

Ex:

$p = [0.2, 0.3, 0.5]$

La valeur vrai est au milieu, sur 0.3
Notre gradient analytique donnerait le r√©sultat : 

$df = [0.2, -0.7, 0.5]$

La valeur $-0.7$ nous indique que accro√Ætre cette classe (la bonne) devrait diminuer la loss $L_i$.

```python
def grad(probs, y, num_examples):
    """
    Compute the gradient on probs
    return : dscores
    """
    dscores = probs
    dscores[range(num_examples), y] -= 1
    dscores /= num_examples
    return dscores


dscores = grad(probs, Y_batch, batch_size)
dscores
```

## Backpropagation

```python
def backpropagation(hidden_layer, dscores, W2, W, X):
    """
    backpropate the gradient to the parameters
    return : dW,db,dW2,db2
    """
    # first backprop into parameters W2 and b2
    dW2 = ponderation(hidden_layer.T, dscores, b=0)
    db2 = np.sum(dscores, axis=0, keepdims=True)

    # next backprop into hidden layer to get the gradient
    dhidden = ponderation(dscores, W2.T, b=0)

    # backprop the ReLU non-linearity (grad of relu) ==> 1 if x > 0 else 0
    dhidden[hidden_layer <= 0] = 0
    # finally into W,b
    dW = ponderation(X.T, dhidden, b=0)
    db = np.sum(dhidden, axis=0, keepdims=True)

    return dW, db, dW2, db2


dW, db, dW2, db2 = backpropagation(hidden_layer,
                                   dscores,
                                   W2,
                                   W,
                                   X_batch)
dW.shape, db.shape
```

## Mises √† jour des poids et biais

```python
def updates(W, b, W2, b2, dW, db, dW2, db2, learning_rate):
    """
    updates the weights
    return : W,b,W2,b2
    """
    W += -learning_rate * dW
    b += -learning_rate * db
    W2 += -learning_rate * dW2
    b2 += -learning_rate * db2

    return W, b, W2, b2


# updates
W, b, W2, b2 = updates(W, b, W2, b2, dW, db, dW2, db2, learning_rate)
```

```python
epochs = 200

train_loss = np.zeros(epochs)
train_accuracy = np.zeros(epochs)
test_loss = np.zeros(epochs)
test_accuracy = np.zeros(epochs)

W = 0.01 * np.random.randn(input_dim, neurons)
b = np.zeros((1, neurons))
W2 = 0.01 * np.random.randn(neurons, ouput_dim)
b2 = np.zeros((1, ouput_dim))

for i in tqdm(range(epochs)):

    X_batch, Y_batch = generate_batch(X_train,
                                      Y_train,
                                      batch_size)

    X_test_batch, Y_test_batch = generate_batch(X_test,
                                                Y_test,
                                                batch_size)

    # 1st layer + relu
    hidden_layer = relu(ponderation(X_batch, W, b))

    # 2nd layer
    scores = ponderation(hidden_layer, W2, b2)

    # softmax
    probs = softmax(scores, eps)

    # crossentropy loss
    loss = cross_entropy(probs, Y_batch, batch_size, eps)

    # logging loss
    train_loss[i] = loss

    # logging accuracy
    train_accuracy[i] = get_accuracy(probs, Y_batch)

    # test set logging
    hidden_layer_test = relu(ponderation(X_test_batch, W, b))
    scores_test = ponderation(hidden_layer_test, W2, b2)
    probs_test = softmax(scores_test, eps)
    loss_test = cross_entropy(probs_test,
                              Y_test_batch,
                              batch_size,
                              eps)

    test_accuracy[i] = get_accuracy(probs_test, Y_test_batch)
    test_loss[i] = loss_test

    # gradient
    dscores = grad(probs, Y_batch, batch_size)

    # backpropagation
    dW, db, dW2, db2 = backpropagation(hidden_layer,
                                       dscores,
                                       W2,
                                       W,
                                       X_batch)

    # updates
    W, b, W2, b2 = updates(W,
                           b,
                           W2,
                           b2,
                           dW,
                           db,
                           dW2,
                           db2,
                           learning_rate)

    if i % (epochs/20) == 0:
        print(
            f"Epoch {i: <4}  train_loss : {round(train_loss[i],4): <9}  test_loss : {round(test_loss[i],4): <9}  train_accuracy : {round(train_accuracy[i],4): <9}  test_accuracy : {round(test_accuracy[i],4): <9}")
```

```python
FP = confusion_matrix.sum(axis=0) - np.diag(confusion_matrix)
FN = confusion_matrix.sum(axis=1) - np.diag(confusion_matrix)
TP = np.diag(confusion_matrix)
TN = confusion_matrix.sum() - (FP + FN + TP)

# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
# Specificity or true negative rate
TNR = TN/(TN+FP)
# Precision or positive predictive value
PPV = TP/(TP+FP)
# Negative predictive value
NPV = TN/(TN+FN)
# Fall out or false positive rate
FPR = FP/(FP+TN)
# False negative rate
FNR = FN/(TP+FN)
# False discovery rate
FDR = FP/(TP+FP)

# Overall accuracy
ACC = (TP+TN)/(TP+FP+FN+TN)
```
